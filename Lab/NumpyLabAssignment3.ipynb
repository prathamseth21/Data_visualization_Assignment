```python
#Ques:1 Calculate the total revenue generated by two product categories in a store.

import numpy as np

category1_revenue = np.array([500,600,700,550])
category2_revenue = np.array([450,700,800,600])

sum1 = np.sum(category1_revenue)
sum2 = np.sum(category2_revenue)

total_sum = sum1 + sum2
print('total sum equals to : ',total_sum)
```

```python
#Ques:2 Calculate the profit made by a company.

import numpy as np

revenue = np.array([10000,12000,11000,10500])
expenses = np.array([4000,5000,4500,4800])

profit = revenue - expenses

print(profit)

print('Total Profit equals to:',np.sum(profit))
```

```python
#Ques:-4 Total cost of items in shopping cart considering the quantity and price.

quantity = np.array([2, 3, 4, 1])
price_per_item = np.array([10.0, 5.0, 8.0, 12.0])

total_cost_per_item = quantity * price_per_item
total_cost = np.sum(total_cost_per_item)

print("Total cost for each item:", total_cost_per_item)
print("Total cost of the shopping cart:", total_cost)
```
